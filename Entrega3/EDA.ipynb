## Limpieza de landmarks

### Cara

Para el proposito de este modelo, solamente es necesario saber la posicion general de la cabeza, por lo tanto, no es necesario tener la informaciÃ³n de los 10 landmarks extra que toma media pipe de la cara:

```python
face_landmarks = [f'landmark_{i}' for i in range(1,11)]
full_face_landmarks = [f'landmark_{i}_x' for i in range(1,11)] + [f'landmark_{i}_y' for i in range(1,11)] + [f'landmark_{i}_z' for i in range(1,11)] + [f'landmark_{i}_vis' for i in range(1,11)]

data = data.drop(columns=full_face_landmarks)
data.describe()
```

### Distancia de hombros

Para detectar de manera mas certera acciones como caminar y girar, se calcula la distancia entre los hombros

```python
distance_soulders = np.sqrt((data['landmark_11_x'] - data['landmark_12_x'])**2 + (data['landmark_11_y'] - data['landmark_12_y'])**2 + (data['landmark_11_z'] - data['landmark_12_z'])**2)
data['distance_soulders'] = distance_soulders
```

### Manos

```python
handsLandmarks = [f'landmark_{i}' for i in range(17, 23)]
full_hands_landmarks = [f'landmark_{i}_x' for i in range(17, 23)] + [f'landmark_{i}_y' for i in range(17, 23)] + [f'landmark_{i}_z' for i in range(17, 23)] + [f'landmark_{i}_vis' for i in range(17, 23)]
data.drop(columns=full_hands_landmarks, inplace=True)
data.head()
```

### Distancia entre landmarks de cadera

Para diferenciar mejor entre caminar y girar, al igual que con los hombros, se obtiene la distancia entre caderas.

```python
hipsDistance = np.sqrt((data['landmark_23_x'] - data['landmark_24_x'])**2 + (data['landmark_23_y'] - data['landmark_24_y'])**2 + (data['landmark_23_z'] - data['landmark_24_z'])**2)
data['hipsDistance'] = hipsDistance
```

# AgrupaciÃ³n

Los datos fueron agrupados de a 10, cada grupo tiene redundancia con el anterior de todos menos su primer y ultimo frame, se decidiÃ³ agrupar para poder detectar el cambio en el movimiento y no las poses especificas que se verian con cada frame.

```python
result_data = []

df = data.copy()
labels = df['labels'].to_numpy()
results_bundle = deque(maxlen=10)
df = df.drop(columns=['labels'])
ardf = df.to_numpy()
result_array = []
for frame in range(len(ardf)):
    if len(results_bundle) == 10:
        new_row = np.array([])
        labels_row = np.array([])
        for n in results_bundle:
            new_row = np.append(new_row, n)
            labels_row = np.append(labels_row, labels[frame])
        if len(np.unique(labels_row)) == 1:
            new_row = np.append(new_row, labels_row[0])
            result_array.append(new_row)

        results_bundle.popleft()
    results_bundle.append(ardf[frame])

    
    
result_df = pd.DataFrame(result_array)

result_df
```

Con el proposito de darle al modelo las herramientas necesarias para encontrar los patrones de cambio para cada movimiento, se decidio calcular el cambio promedio en diferentes ejes de diferentes apendices, con el proposito de darle a cada grupo un dato con el cual basarse para el analisis.
